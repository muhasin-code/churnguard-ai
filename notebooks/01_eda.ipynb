## 3. Univariate Analysis

### For numerical features:

```python
numerical_columns = df.select_dtypes(include=["int64", "float64"]).columns

for column in numerical_columns:
    plt.figure(figsize=(10, 4))

    # Distribute
    plt.subplot(1, 2, 1)
    df[column].hist(bins=50)
    plt.title(f"{column} Distribution")

    # Boxplot (detect outliers)
    plt.subplot(1, 2, 2)
    df.boxplot(column=column)
    plt.tight_layout()
    plt.show()

    # Statistics
    print(f"\n{column} Statistics")
    print(df[column].describe())
```

### For categorical features

```python
categorical_columns = df.select_dtypes(include=["object"]).columns.drop("CustomerID")

for column in categorical_columns:
    print(f"\n{column} Value Counts:")
    print(df[column].value_counts())

    # Bar plot
    df[column].value_counts().plot(kind='bar')
    plt.title(f"{column} Distribution")
    plt.show()
```

## 4. Bivariate Analysis

### Numerical vs Target:

```python
for column in numerical_columns:
    # Grouped Statistics
    print(f"{column} by Churn:")
    print(df.groupby("Churn")[column].describe())

    # Violin plot
    sns.violinplot(data=df, x="Churn", y=column)
    plt.title(f"{column} vs Churn")
    plt.show()
```

### Categorical vs Target

```python
for column in categorical_columns:
    if column == "Churn":
        continue

    # Cross-tabulation
    ct = pd.crosstab(df[column], df["Churn"], normalize="index")
    print(f"\n{column} vs Churn:")
    print(ct)

    # Stacked bar chart
    ct.plot(kind='bar', stacked=True)
    plt.title(f"{column} vs Churn")
    plt.show()
```

### Churn Rate by Categorical Features

```python
# Calculate churn rate by category
for column in ["ContractType", "InternetService", "PaymentMethod"]:
    churn_rate = df.groupby(column)['Churn'].apply(
        lambda x: (x == 'Yes').mean() * 100
    ).sort_values(ascending=False)

    print(f"\n{column} Churn Rates:")
    print(churn_rate)
```

### Outlier Quantification

```python
# Quantify outliers using IQR method
def count_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 - 1.5 * IQR

    outliers = df[(df[column] < lower_bound) | (df[column] < upper_bound)]

    print(f"\n{column}")
    print(f" Outliers: {len(outliers)} ({len(outliers) / len(df) * 100:.2f}%)")
    print(f" Range: [{lower_bound:.2f}, {upper_bound:.2f}]")

    return outliers

# Check for outliers
for column in ['MonthlyCharges', 'TotalCharges', 'CallMinutes']:
    outliers = count_outliers(df, column)
```

## Feature Engineering Ideas

Based on EDA insights:

### 1. Tenure Buckets
- **Rationale:** Clear churn pattern - new customers (<6 months) churn more
- **Implementation:** `pd.cut(tenure, bins=[0, 6, 12, 24, 60, 100])`
- **Expected impact:** Capture non-linear tenure-churn relationship

### 2. Price-Per-Service Ratio
- **Rationale:** Customers paying high for few services might feel poor value
- **Implementation:** `monthly_charges / num_services`
- **Expected impact:** Identify value-perception issues

### 3. Contract-Tenure Interaction
- **Rationale:** Month-to-month with high tenure is unusual (exit planning?)
- **Implementation:** `is_monthly * tenure`
- **Expected impact:** Catch customers preparing to leave

### 4. Support Intensity
- **Rationale:** Recent tickets signal dissatisfaction
- **Implementation:** `recent_support_tickets * complaints`
- **Expected impact:** Amplify frustration signals

### 5. Engagement Drop Indicator
- **Rationale:** Low engagement + high price = churn risk
- **Implementation:** `(engagement < 1) & (monthly_charges > 70)`
- **Expected impact:** Catch disengaged high-value customers
